```python
## XPATH

xp = '//*[@id="main_pack"]/div[2]/div[2]/div/div/div[4]/div[4]/ul/li'
reviews = driver.find_elements(By.XPATH, xp)
```

```python
reviews[0]
```

```python
## <테그> 텍스트 </> 데이터 추출
texts = []
for review in reviews:
    text = review.find_element(By.CSS_SELECTOR, 'dd.this_text_stress._btn_writer').get_attribute('innerText')
    texts.append(text)
texts
```

```python
text1 = review.find_element(By.CSS_SELECTOR, 'div.area_text_box').get_attribute('innerText')
text_sp = text1.split("\n")
text_sp[1]
```

```python
def get_content(review):
    condic = {}
    condic['text'] = review.find_element(By.CSS_SELECTOR, 'span.desc._text').get_attribute('innerText').strip()
    condic['id'] = review.find_element(By.CSS_SELECTOR, 'dd.this_text_stress._btn_writer').get_attribute('innerText')
    text1 = review.find_element(By.CSS_SELECTOR, 'div.area_text_box').get_attribute('innerText')
    text_sp = text1.split("\n")
    text_sp[1]
    condic['rat'] = text_sp[1]
    condic['date'] = review.find_element(By.CSS_SELECTOR, 'dd.this_text_normal').get_attribute('innerText').strip()
    
    return condic
```

```python
import pandas as pd
result = []

for content in reviews:
    try:
        result.append(get_content(content))
    except:
        print('error')
        pass
    
df = pd.DataFrame(result)
df
```

```python
df.to_csv('./data/Moive_review_Seouls_Spring.csv', index = False)
```
