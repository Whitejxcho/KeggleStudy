```python
import re

text_sub = []
for text in df['text']:
    rtext = str(text).strip()
    rtext = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》\xa0\n\d+]', '', rtext)
    sptext = rtext.split(' ')
    nw = [w for w in sptext if w not in ['','초','이','등']]
    rtext = ' '.join(nw)
    text_sub.append(rtext)
df['text_sub'] = text_sub
# df
```

```python
## 고유 토큰을 찾기위해 set 구조 활용
uq = set()
for text in df['text_sub']:
    sp = text.split(' ')
    for w in sp:
        uq.add(w)
# uq
```

```python
## 고유 토큰에 고유 번호를 맵핑시킨 딕셔너리 구성
enc_dict = {}
for i,t in enumerate(uq):
    enc_dict[t] = i
enc_dict
```

```python
## 딕셔너리를 활용하여 인코딩
text = df.loc[5,'text_sub']
str_tk = text.split(' ')
enc_tk = [enc_dict[w] for w in str_tk]
enc_tk
```

```python
# 판다스 출력 옵션 설정(무한대 출력)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

def show_tag(text):
    mc_text = mecab.pos(text)
    kw_text = [(t.form, t.tag) for t in kw.tokenize(text)] 
    max_len = max([len(mc_text),len(kw_text)])

    mc_text += [None for i in range(max_len - len(mc_text))]
    kw_text += [None for i in range(max_len - len(kw_text))]

    text_set = {
        'Mecab':mc_text,
        'Kiwi':kw_text
    }
    return pd.DataFrame.from_dict(text_set)
```

```python
from kiwipiepy import Kiwi

store_df = pd.read_csv('./data/Moive_review_Seouls_Spring.csv').dropna()

kw = Kiwi()
use_tags = ['NNP', 'NNG', 'NP', 'VA', 'IC', 'VV', 'MAG']

rtexts = []
for text in store_df['text']:
    kwords = kw.tokenize(text)
    kwords = [w.form for w in kwords if w.tag in use_tags]
    rnew_text = ' '.join(kwords)
    rtexts.append(rnew_text)
    
store_df['text_tag'] = rtexts
store_df.to_csv('./data/Moive_review_Seouls_Spring.csv',index = False)
store_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer 

df = pd.read_csv('./data/Moive_review_Seouls_Spring.csv').dropna()

## max_df : 해당 확률 이상으로 빈도가 많은 토큰을 제외
## min_df: 해당 확률 이하로 빈도가 적은 토큰을 제외
cvec = CountVectorizer(max_df=0.9, min_df=10)
cvec.fit(df['text_tag']) # 토큰생성 (띄어쓰기 기반)
names = cvec.vocabulary_ # 토큰셋 확인
print(len(names))
print(names)
```
